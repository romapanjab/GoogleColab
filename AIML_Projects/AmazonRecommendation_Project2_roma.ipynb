```python
# Clean price and discount columns
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)

# Extract main category
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])

# Unique counts
unique_users = df['user_id'].nunique()
unique_products = df['product_id'].nunique()
unique_reviews = df['review_id'].nunique()

# Convert to numpy
np_unique_users = np.array([unique_users])
np_unique_products = np.array([unique_products])
np_unique_reviews = np.array([unique_reviews])

# Top 5 categories
top_categories = df['main_category'].value_counts().head(5)
np_top_categories = top_categories.to_numpy()

# Price and discount insights
price_range = (df['discounted_price'].min(), df['discounted_price'].max())
avg_discount = df['discount_percentage'].mean()
max_discount = df['discount_percentage'].max()

# Numpy arrays
np_price_range = np.array(price_range)
np_avg_discount = np.array([avg_discount])
np_max_discount = np.array([max_discount])

# Print outputs
print("Unique Users:", np_unique_users)
print("Unique Products:", np_unique_products)
print("Unique Reviews:", np_unique_reviews)
print("\nTop 5 Categories by Product Count:\n", top_categories)
print("\nPrice Range (₹):", np_price_range)
print("Average Discount (%):", np_avg_discount)
print("Maximum Discount (%):", np_max_discount)

# Top 5 Categories
plt.figure(figsize=(8, 5))
top_categories.plot(kind='bar')
plt.title("Top 5 Product Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Price Distribution
plt.figure(figsize=(8, 5))
plt.hist(df['discounted_price'], bins=30, edgecolor='k')
plt.title("Distribution of Discounted Prices")
plt.xlabel("Discounted Price (₹)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Discount Percentage Distribution
plt.figure(figsize=(8, 5))
plt.hist(df['discount_percentage'], bins=30, edgecolor='k', color='orange')
plt.title("Distribution of Discount Percentage")
plt.xlabel("Discount Percentage (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

**Clean and preprocess the data:**
1. Convert prices to numeric
2. Parse categories into hierarchy levels
3. **Normalize** rating scores and count outliers or Create derived features like price_difference, value_for_money_score, weighted ratings.

**Steps needed to achieve goal 3 from above points:**

1. df['rating'].fillna(0)
This replaces missing ratings (NaN) with 0 to avoid errors during normalization.
(Note: You could also use .dropna() or .fillna(df['rating'].mean()) depending on the goal.)

2. **zscore(df['rating'].fillna(0))**
This calculates the z-score of the rating column:

𝑧 =(𝑥−𝜇)/𝜎

where:
x is the individual rating
μ is the mean rating
σ is the standard deviation of the ratings

The result is a normalized score showing how far each rating deviates from the mean (in standard deviations).

3. df['rating_zscore'] = ...
Stores the z-score result in a new column called rating_zscore.

4. **outlier_count = (df['rating_zscore'].abs() > 3).sum()**
Counts how many ratings are more than 3 standard deviations away from the mean.
These are considered statistical outliers in a normal distribution.

```python
from scipy.stats import zscore

# Convert rating and rating_count to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Parse category hierarchy
category_levels = df['category'].str.split('|', expand=True)
category_levels.columns = [f'level_{i+1}' for i in range(category_levels.shape[1])]
df = pd.concat([df, category_levels], axis=1)

# Derived features
df['price_difference'] = df['actual_price'] - df['discounted_price']
df['value_for_money_score'] = df['rating'] / df['discounted_price']
df['weighted_rating'] = df['rating'] * df['rating_count']

# Normalize ratings and identify outliers
df['rating_zscore'] = zscore(df['rating'].fillna(0))
outlier_count = (df['rating_zscore'].abs() > 3).sum()

# Display sample of results
print(df[['product_name', 'discounted_price', 'actual_price', 'price_difference',
          'rating', 'rating_count', 'value_for_money_score', 'weighted_rating',
          'level_1', 'level_2', 'level_3', 'rating_zscore']].head())

# Print outlier count
print("Number of outliers in normalized ratings (z-score > 3):", outlier_count)
```

**Section C: Content-Based Filtering:**
Vectorize product text (about_product + product_name) using:
- TF-IDF or embeddings
- Build a product similarity matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine product_name and about_product for vectorization
df['combined_text'] = df['product_name'].fillna('') + ' ' + df['about_product'].fillna('')

# Initialize and apply TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = tfidf.fit_transform(df['combined_text'])
print(tfidf_matrix)

# Compute cosine similarity matrix
product_similarity = cosine_similarity(tfidf_matrix)

#✅ product_similarity
#A NumPy array of shape (n_products, n_products)
#Each value represents the cosine similarity between a pair of products (from 0 to 1)
#✅ pd.DataFrame(...)
#Wraps the similarity matrix into a Pandas DataFrame for better usability
#✅ index=df['product_name']
#Assigns product names as row labels
# This means each row corresponds to a product
#✅ columns=df['product_name']
#Assigns product names as column labels
#This means each column also corresponds to a product

# Convert to a pandas DataFrame for easy access
similarity_df = pd.DataFrame(product_similarity, index=df['product_name'], columns=df['product_name'])

# Preview sample of similarity matrix
print(similarity_df.iloc[:5, :5])
```

 Explanation:
1. TfidfVectorizer(stop_words='english', max_features=1000)
- Initializes a TF-IDF Vectorizer, which converts text data into numerical features based on word importance.
- stop_words='english': Removes common English words (like the, is, and) that don't carry meaningful information.
- max_features=1000: Only the top 1,000 most important words (by frequency and uniqueness) across the dataset will be kept as features.

2. tfidf.fit_transform(df['combined_text'])
- fit: Learns the vocabulary and IDF (inverse document frequency) from all the text.
- transform: Converts each product's combined text into a vector of 1,000 TF-IDF features.

The result is a sparse matrix (tfidf_matrix) of shape (num_products, 1000).
Each row in tfidf_matrix represents a product, and each column corresponds to a term in the vocabulary.

💡 Use Case:
You use this matrix to:
- Measure similarity between products
- Build recommendation systems
- Perform clustering or dimensionality reduction

**Next part: **
Recommend top 5 similar products to:
- A new product with no reviews
- A product with high user dropout (bad ratings)

```python
# -------- Recommendation for a New Product (simulate using 1st row) --------
new_product_text = df.iloc[0]['combined_text']

#tfidf.transform(...):
#This applies the existing TF-IDF vocabulary and IDF weights (learned earlier using .fit_transform(...) on the main dataset).
#It transforms the new text into a sparse vector of TF-IDF scores corresponding to the learned terms.
new_vector = tfidf.transform([new_product_text])


#cosine_similarity(new_vector, tfidf_matrix): Computes cosine similarity between:
#ew_vector: a 1-row TF-IDF vector for the new product
#tfidf_matrix: TF-IDF vectors of all existing products
#Output is a 2D array of shape (1, N), where N = number of existing products.
#.flatten(): This method converts a 2D array into a 1D array.
#This makes it easier to sort and index the similarity scores with: new_similarities.argsort()

new_similarities = cosine_similarity(new_vector, tfidf_matrix).flatten()

# Get top 5 similar products excluding itself

#new_similarities.argsort(): Returns the indices that would sort the array of similarity scores in ascending order.
# So the least similar product comes first, and the most similar product comes last.
#[::-1]: Reverses the sorted index array to get descending order (most similar first).
#df.iloc[...]: Uses the resulting indices to select rows from the DataFrame.
#✅ [["product_name", "rating", "rating_count"]]: Selects only the columns we want to show in the output.

top5_similar_to_new = df.iloc[new_similarities.argsort()[::-1][1:6]][['product_name', 'rating', 'rating_count']]
print("Top 5 Recommendations for New Product:\n", top5_similar_to_new)

# -------- Recommendation for Low-Rated Product --------
# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Find index of the lowest-rated product
low_rated_product_index = df['rating'].idxmin()
low_rated_vector = tfidf_matrix[low_rated_product_index]

# Compute similarity and get top 5 similar products
low_rating_similarities = cosine_similarity(low_rated_vector, tfidf_matrix).flatten()
top5_similar_to_low_rated = df.iloc[low_rating_similarities.argsort()[::-1][1:6]][['product_name', 'rating', 'rating_count']]
print("\nTop 5 Recommendations for Low-Rated Product:\n", top5_similar_to_low_rated)
```

**Next section:** Add category, price, and discount to enhance content vectors.

Evaluate recommendations:
How diverse and relevant are the content-based results?

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.sparse import hstack

# Combine text fields with category
df['combined_text'] = (
    df['product_name'].fillna('') + ' ' +
    df['about_product'].fillna('') + ' ' +
    df['category'].fillna('')
)

# Vectorize text using TF-IDF
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = tfidf.fit_transform(df['combined_text'])

# Normalize numeric features (price and discount)
numeric_features = df[['actual_price', 'discount_percentage']].fillna(0)

scaler = MinMaxScaler()
#What is MinMaxScaler()?: It’s a normalization technique that transforms numeric features into a common scale, typically [0, 1].
#Why Use It?: To ensure that price, discount, and other numeric features are on comparable scales
#Prevents larger numbers (like ₹79,990) from overpowering smaller-scale features (like TF-IDF values)
#Crucial when combining text vectors and numeric data into one model

numeric_scaled = scaler.fit_transform(numeric_features)

# Combine text and numeric features
enhanced_matrix = hstack([tfidf_matrix, numeric_scaled]).tocsr()

# Simulate a new product (use first product as a placeholder)
new_vector = enhanced_matrix[0]
similarities = cosine_similarity(new_vector, enhanced_matrix).flatten()

# Get top 5 similar products excluding itself
top5_indices = similarities.argsort()[::-1][1:6]
top5_enhanced = df.iloc[top5_indices][[
    'product_name', 'category', 'actual_price', 'discount_percentage', 'rating', 'rating_count']]

print("Top 5 Enhanced Recommendations:\n", top5_enhanced)

# Evaluate Diversity and Relevance
diversity_score = top5_enhanced['category'].nunique()
relevance_score = pd.to_numeric(top5_enhanced['rating'], errors='coerce').mean()

print("\nDiversity Score (Unique Categories):", diversity_score)
print("Average Relevance Score (Ratings):", relevance_score)
```

Next Section: Collaborative Filtering:
Create a user-item matrix using user_id, product_id, and rating.

Apply: User-User Collaborative Filtering (cosine similarity) OR Item-Item Collaborative Filtering (cosine or Pearson).

Recommend top 5 unseen products per user

```python
from sklearn.metrics.pairwise import cosine_similarity

# Convert 'rating' column to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Create the User-Item matrix (rows: users, columns: products)
user_item_matrix = df.pivot_table(index='user_id', columns='product_id', values='rating', aggfunc='mean')

# Fill missing values with 0 (needed for cosine similarity calculation)
user_item_filled = user_item_matrix.fillna(0)

# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_filled)
#converting to a pandas dataframe for the ease of use of sorting, filtering, accessing specific user pairs
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

# Recommend top 5 unseen products for a given user
def recommend_top_5(user_id, user_item_matrix, similarity_df):
    # Find top 5 similar users (excluding the user itself)
    similar_users = similarity_df[user_id].sort_values(ascending=False).drop(user_id).head(5)

    # Weighted sum of ratings from similar users
    weighted_ratings = user_item_matrix.loc[similar_users.index].T.dot(similar_users)

    # Normalize by similarity total
    sim_sum = similar_users.sum()
    scores = weighted_ratings / sim_sum

    # Filter out products already rated by the target user
    seen_products = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id].notna()].index
    recommendations = scores.drop(index=seen_products).sort_values(ascending=False).head(5)

    return recommendations

# Example: Run for a sample user
sample_user_id = user_item_matrix.index[0]
top5_recommendations = recommend_top_5(sample_user_id, user_item_matrix, user_similarity_df)


#The Goal: You already have the top 5 recommended product IDs with their predicted scores (top5_recommendations), and now you want to get:
#Their product names In the same order as in the recommendation list
#🧠 Step-by-Step Breakdown:
#✅ 1. df[df['product_id'].isin(top5_recommendations.index)]
#Filters the original dataset to include only the top 5 recommended products
#✅ 2. [['product_id', 'product_name']]
#Selects only the columns we care about for display
#✅ 3. .drop_duplicates()
#Removes duplicate rows, in case the same product appears multiple times in the dataset (e.g., multiple reviews)
#✅ 4. .set_index('product_id')
#Sets product_id as the index, so you can use .loc[...] with it
#✅ 5. .loc[top5_recommendations.index]
#Reorders the DataFrame to match the order of top5_recommendations, because isin() doesn’t preserve order
#✅ Final Output:
#You get a DataFrame of the top 5 recommended products: Indexed by product_id, Includes the matching product_name
#In the same order as predicted by the model
# Display product names for the recommended product IDs
top5_df = df[df['product_id'].isin(top5_recommendations.index)][['product_id', 'product_name']] \
            .drop_duplicates().set_index('product_id').loc[top5_recommendations.index]

top5_df['Predicted_Score'] = top5_recommendations.values

print("Top 5 Product Recommendations for User:", sample_user_id)
print(top5_df)
```

Next Section

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler

# User-Item rating matrix
cf_matrix = user_item_matrix.fillna(0)

# Collaborative Filtering (user-user)
cf_similarity = cosine_similarity(cf_matrix)
cf_similarity_df = pd.DataFrame(cf_similarity, index=cf_matrix.index, columns=cf_matrix.index)

# Content-Based: TF-IDF + price + discount
df['combined_text'] = df['product_name'].fillna('') + ' ' + df['about_product'].fillna('') + ' ' + df['category'].fillna('')
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = tfidf.fit_transform(df['combined_text'])

scaler = MinMaxScaler()
numeric_scaled = scaler.fit_transform(df[['actual_price', 'discount_percentage']].fillna(0))
content_matrix = hstack([tfidf_matrix, numeric_scaled]).tocsr()
content_similarity = cosine_similarity(content_matrix)

# Mapping for indexing
#maps product ids to row index
product_index_map = dict(zip(df['product_id'], df.index))
#the reverse means helps to get the product id given the row index
reverse_product_map = {v: k for k, v in product_index_map.items()}

# Hybrid recommender function
def hybrid_recommend_for_user(user_id, top_n=5):
    if user_id not in user_item_matrix.index:
        # Cold-start product fallback

        #content_similarity is a matrix where each value shows how similar two products are.
        #We average the similarity of each product to all others (across columns).
        #ravel() flattens it into a 1D array → one average score per product.
        avg_content_scores = content_similarity.mean(axis=0).ravel()
        top_indexes = [i for i in avg_content_scores.argsort()[::-1] if i in reverse_product_map][:top_n]
        product_ids = [reverse_product_map[i] for i in top_indexes]
        return df.set_index('product_id').loc[product_ids][['product_name', 'rating', 'rating_count']]

    # CF Scores from top similar users
    similar_users = cf_similarity_df[user_id].sort_values(ascending=False).drop(user_id).head(5)
    weighted_ratings = user_item_matrix.loc[similar_users.index].T.dot(similar_users)
    cf_scores = weighted_ratings / similar_users.sum()

    # Products already rated
    seen_products = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id].notna()].index
    seen_indexes = [product_index_map[pid] for pid in seen_products if pid in product_index_map]

    #If the user hasn’t rated any products: seen_indexes will be an empty list
    #That means we can’t compute collaborative similarity
    #So we fallback to content-based recommendations (for cold-start users)
    if not seen_indexes:
        return hybrid_recommend_for_user('cold_user', top_n)

    # Content scores from similar items
    avg_sim = content_similarity[seen_indexes].mean(axis=0).ravel()
    content_scores = pd.Series(avg_sim, index=df['product_id'])

    # Score fusion
    combined = 0.6 * cf_scores + 0.4 * content_scores
    combined = combined.drop(index=seen_products, errors='ignore').sort_values(ascending=False).head(top_n)

    return df.set_index('product_id').loc[combined.index][['product_name', 'rating', 'rating_count']]

# Identify cold-start user
user_rating_counts = df['user_id'].value_counts()
#.index[0] Grabs the first user ID from the filtered result.
cold_user_id = user_rating_counts[user_rating_counts == 1].index[0]

# Get recommendations
cold_user_recs = hybrid_recommend_for_user(cold_user_id)
cold_product_recs = hybrid_recommend_for_user('cold_user')

# Display
print("\nCold-Start User Recommendations:")
print(cold_user_recs)

print("\nCold-Start Product Recommendations:")
print(cold_product_recs)
```

Line by Line explanation of code:

scaler = MinMaxScaler()
Initializes a MinMaxScaler from sklearn.
This scales numeric features to the range [0, 1].
Helps ensure all features (price, discount) are on the same scale as TF-IDF values.

🔢 numeric_scaled = scaler.fit_transform(df[['actual_price', 'discount_percentage']].fillna(0))
Selects actual_price and discount_percentage columns.
Fills missing values with 0.
Normalizes both columns into values between 0 and 1.
numeric_scaled becomes a NumPy array with shape (n_products, 2).

🔗 content_matrix = hstack([tfidf_matrix, numeric_scaled]).tocsr()
hstack(...) horizontally concatenates:
- tfidf_matrix (text vector for each product)
- numeric_scaled (normalized price and discount)
This creates a new content vector for each product that includes:
- Textual info (product name, description, category)
- Numeric info (price, discount)

.tocsr() converts it to CSR format, a sparse matrix format efficient for similarity computation.

📏 content_similarity = cosine_similarity(content_matrix)
Computes cosine similarity between all product content vectors.
Each value in the matrix tells how similar two products are based on both:
their text
and their numeric properties

📌 Final Result:
Now you have:

content_similarity[i][j] = similarity score between product_i and product_j
This is used to:
Recommend products similar to a given product
Find alternatives based on features + price + discount

Why Do This Hybrid strategy?
To evaluate how well your system performs under real-world limitations:

🧊 New user joins → no data → fallback to content

🧊 User with 1-2 reviews → use hybrid with limited CF

🚫 Avoid blank screens or irrelevant products
